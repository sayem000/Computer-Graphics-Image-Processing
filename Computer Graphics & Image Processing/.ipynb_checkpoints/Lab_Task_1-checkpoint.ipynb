# Bresenham Line Drawing Algorithm Implementation

```python
import matplotlib.pyplot as plt
```

```python
def bresenham_algo(x0, y0, x1, y1):
    steep = abs(y1 - y0) > abs(x1 - x0)

    if steep:
        x0, y0 = y0, x0
        x1, y1 = y1, x1
    points = []
    dx = abs(x1 - x0)
    dy = abs(y1 - y0)
    x, y = x0, y0

    p = 2*dy-dx

    while True:
        points.append((y, x) if steep else (x, y))

        if x == x1 and y == y1:
            break
        if p < 0:
            x = x + 1
            y = y
            p = p + 2*dy
        else:
            x = x + 1
            y = y + 1
            p = p + 2*dy - 2*dx
    return points
```

# Case 01

```python
x0, y0 = 1, 1
x1, y1 = 8, 4
line_points = bresenham_algo(x0, y0, x1, y1)
```

```python
x_values, y_values = zip(*line_points)
plt.plot(x_values, y_values, marker='^', color='red')
plt.grid(True)
plt.show()
```
